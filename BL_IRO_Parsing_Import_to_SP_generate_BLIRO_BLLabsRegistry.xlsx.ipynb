```python
filename = "BL_IRO_export.csv"  # Rename the CSV file received from the BL Repository to this name (without the date)

df = pd.read_csv(filename, low_memory=False)
```

```python
# slice the dataframe to get only the Datasets

datasets_df = df[df.model == "Dataset"] 
```

```python
def authors(rec):
    """
    This function concatenates the contributors (up to 143 contributors, with 9 fields each) and displays them according to the MHRA style.
    """
    
    # Authors (AKA Contributors) -- How do I cite sources with multiple authors in MHRA referencing?
    # https://www.scribbr.co.uk/faqs/sources-with-multiple-authors-in-mhra/

    i = 1

    while datasets_df.iloc[rec]["creator_family_name_" + str(i)]:
      i += 1
    if i == 1:
        #print("No authors?")
        return ''
    else:
        if i == 2:
            return datasets_df.iloc[rec]["creator_family_name_" + str(1)] + ", " + datasets_df.iloc[rec]["creator_given_name_" + str(1)]
        else: 
            if i == 3:
                return datasets_df.iloc[rec]["creator_family_name_" + str(1)] + ", " + datasets_df.iloc[rec]["creator_given_name_" + str(1)] + ", and " + datasets_df.iloc[rec]["creator_given_name_" + str(2)] + " " + datasets_df.iloc[rec]["creator_family_name_" + str(2)]
            else:
                if i == 4:
                    return datasets_df.iloc[rec]["creator_family_name_" + str(1)] + ", " + datasets_df.iloc[rec]["creator_given_name_" + str(1)] + ", " + datasets_df.iloc[rec]["creator_given_name_" + str(2)] + " " + datasets_df.iloc[rec]["creator_family_name_" + str(2)] + " and " + datasets_df.iloc[rec]["creator_given_name_" + str(3)] + " " + datasets_df.iloc[rec]["creator_family_name_" + str(3)]
                else:
                    return datasets_df.iloc[rec]["creator_family_name_" + str(1)] + ", " + datasets_df.iloc[rec]["creator_given_name_" + str(1)] + ", and others"
```

```python
def contributors(rec):
    """
    This function concatenates the contributors (up to 143 contributors, with 9 fields each) and displays them according to the MHRA style.
    """
    
    # Authors (AKA Contributors) -- How do I cite sources with multiple authors in MHRA referencing?
    # https://www.scribbr.co.uk/faqs/sources-with-multiple-authors-in-mhra/

    i = 1

    while datasets_df.iloc[rec]["contributor_family_name_" + str(i)]:
      i += 1
    if i == 1:
        #print("No authors?")
        return ''
    else:
        if i == 2:
            return datasets_df.iloc[rec]["contributor_family_name_" + str(1)] + ", " + datasets_df.iloc[rec]["contributor_given_name_" + str(1)]
        else: 
            if i == 3:
                return datasets_df.iloc[rec]["contributor_family_name_" + str(1)] + ", " + datasets_df.iloc[rec]["contributor_given_name_" + str(1)] + ", and " + datasets_df.iloc[rec]["contributor_given_name_" + str(2)] + " " + datasets_df.iloc[rec]["contributor_family_name_" + str(2)]
            else:
                if i == 4:
                    return datasets_df.iloc[rec]["contributor_family_name_" + str(1)] + ", " + datasets_df.iloc[rec]["contributor_given_name_" + str(1)] + ", " + datasets_df.iloc[rec]["contributor_given_name_" + str(2)] + " " + datasets_df.iloc[rec]["contributor_family_name_" + str(2)] + " and " + datasets_df.iloc[rec]["contributor_given_name_" + str(3)] + " " + datasets_df.iloc[rec]["contributor_family_name_" + str(3)]
                else:
                    return datasets_df.iloc[rec]["contributor_family_name_" + str(1)] + ", " + datasets_df.iloc[rec]["contributor_given_name_" + str(1)] + ", and others"
```

```python
def Keywords(rec):
    
    """
    This function concatenates the keywords, grouping them with a ', '.
    """
    
    i = 1
    Keywords = ''

    while datasets_df.iloc[rec]["keyword_" + str(i)]:
        if i == 1:
            Keywords = datasets_df.iloc[rec]["keyword_" + str(i)]
        else:
            Keywords = Keywords + ", " + datasets_df.iloc[rec]["keyword_" + str(i)]
        i += 1
    
    return Keywords
```

```python
def files(rec):
    
    """
    This function concatenates the files found, grouping them with a ', '.
    """
    
    i = 1
    files = ''
    try:
        while datasets_df.iloc[rec]["file_" + str(i)]:
            if i == 1:
                files = datasets_df.iloc[rec]["file_" + str(i)]
            else:
                files = files + ", " + datasets_df.iloc[rec]["file_" + str(i)]
            i += 1
    except:
        return files    
    return files
```

```python
def Size(rec):

    Size = ''
    end_prev = 0
    
    regex = r"\b(?=(\W+\w+\s+GB|\S+\w+\s+GB|\W+\w+\s+MB|\S+\w+\s+MB|\W+\w+\s+KB|\S+\w+\s+KB)\b)"

    try:
        test_str = datasets_df.loc[rec]["additional_information"]

        matches = re.finditer(regex, test_str, re.MULTILINE)

        for matchNum, match in enumerate(matches, start=1):

            # print ("Match {matchNum} was found at {start}-{end}: {match}".format(matchNum = matchNum, start = match.start(), end = match.end(), match = match.group()))

            for groupNum in range(0, len(match.groups())):
                groupNum = groupNum + 1

                if match.group(groupNum)[0] != ".":
                    start = match.start(groupNum)
                    end = match.end(groupNum)
                    group = match.group(groupNum)

                    if start >= end_prev:
                        if Size == '':
                            Size = str(group).strip();
                        else:
                            Size = Size + ", " + str(group).strip();
                        end_prev = end

        return Size

    except:
         return ''
```
